```python
import pandas as pd
import numpy as np
```

```python
pop3 = pd.read_csv('qgis/최종 3 고령인구.csv',encoding='cp949')
```

```python
pop3.keys()
```

```python
layer = list(pop3['layer'].unique())
layer
```

```python
pop3['layer_2'].unique()
```

```python
layer[0]
```

```python
pop3_drugstore = pop3[pop3['layer']==layer[0]]
pop3_welfare_DamYang = pop3[pop3['layer']==layer[1]]
pop3_healthB = pop3[pop3['layer']==layer[2]]
pop3_healthC = pop3[pop3['layer']==layer[3]]
pop3_welfare_HamPyeong = pop3[pop3['layer']==layer[4]]
pop3_welfare_ShinAhn = pop3[pop3['layer']==layer[5]]
pop3_healthA = pop3[pop3['layer']==layer[6]]
```

```python
pop3_drugstore = pop3_drugstore[['약국명_y','layer_2','좌표정보(X)','좌표정보(Y)','gid','val']]
pop3_drugstore.rename(columns={'약국명_y':'name',
                               '좌표정보(X)':'Longitude',
                               '좌표정보(Y)':'Latitude'}
                      ,inplace=True)
pop3_drugstore
```

```python
pop3_welfare_DamYang = pop3_welfare_DamYang[['회관명','layer_2','Longitude','Latitude','gid','val']]
pop3_welfare_DamYang.rename(columns={'회관명':'name'},inplace=True)
pop3_welfare_DamYang
```

```python
pop3_welfare_DamYang.info()
```

```python
pop3_welfare_HamPyeong = pop3_welfare_HamPyeong[['경로당명','layer_2','Longitude','Latitude','gid','val']]
pop3_welfare_HamPyeong.rename(columns={'경로당명':'name'},inplace=True)
pop3_welfare_HamPyeong
```

```python
pop3_welfare_ShinAhn = pop3_welfare_ShinAhn[['마을회관명','layer_2','Longitude','Latitude','gid','val']]
pop3_welfare_ShinAhn.rename(columns={'마을회관명':'name'},inplace=True)
pop3_welfare_ShinAhn
```

```python
pop3_healthA = pop3_healthA[['보건소명','layer_2','Longitude','Latitude','gid','val']]
pop3_healthB = pop3_healthB[['기관명','layer_2','Longitude','Latitude','gid','val']]
pop3_healthC = pop3_healthC[['기관명','layer_2','Longitude','Latitude','gid','val']]
pop3_healthA.rename(columns={'보건소명':'name'},inplace=True)
pop3_healthB.rename(columns={'기관명':'name'},inplace=True)
pop3_healthC.rename(columns={'기관명':'name'},inplace=True)
```

```python
pop3 = pd.concat([pop3_drugstore, pop3_welfare_DamYang,
                  pop3_welfare_ShinAhn,pop3_welfare_HamPyeong,
                  pop3_healthA,pop3_healthB,pop3_healthC],
                 axis=0,
                 ignore_index = True)
```

```python
#pop3['index'] = pop3['name'] + str(pop3['Longitude'])
```

```python
pop3
```

```python
pop3_ = pop3.groupby('name').sum('val')
```

```python
pop3_
```

```python
pop3_.to_csv('qgis/고령인구3km.csv',encoding='cp949')
```
