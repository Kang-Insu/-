```python
import pandas as pd
```

```python
jeonnam = pd.read_csv('전라남도.csv')
jeonnam
```

```python
pharm = pd.read_csv('전국약국현황(좌표값 있음).csv',encoding='cp949')
pharm
```

```python
pharm.info()
```

```python
pharm = pharm[['요양기관명','시도명','전화번호','주소','좌표정보(X)','좌표정보(Y)']]
pharm = pharm[pharm['시도명']=='전남']
pharm = pharm.rename(columns={'요양기관명':'약국명'})
pharm
```

```python
jeonnam = jeonnam.rename(columns={'대표전화':'전화번호'})
jeonnam
```

```python
pharm_jn = pd.merge(pharm, jeonnam, on = '전화번호', how='right')
pharm_jn
```

```python
pharm_jn.info()
```

```python
pharm_jn = pharm_jn[['약국명_y','주소_y','전화번호','좌표정보(X)','좌표정보(Y)','월요일 운영','화요일 운영','수요일 운영','목요일 운영','금요일 운영',
                    '토요일 운영','일요일 운영','공휴일 운영']]
pharm_jn
```

```python
pharm_jn.to_csv('전남약국데이터.csv',encoding='cp949')
```

```python
jn_pharm = pd.read_csv('전남약국.csv')
jn_pharm
```

```python
jn = jn_pharm[jn_pharm['공휴일 운영']!='-']
jeonnam_ed = jn[jn['일요일 운영']!='-']
jeonnam_ed = jeonnam_ed[jeonnam_ed['토요일 운영']!='-']
jeonnam_ed = jeonnam_ed[jeonnam_ed['토요일 운영']!='-']
jeonnam_ed = jeonnam_ed[jeonnam_ed['월요일 운영']!='-']
jeonnam_ed = jeonnam_ed[jeonnam_ed['화요일 운영']!='-']
jeonnam_ed = jeonnam_ed[jeonnam_ed['수요일 운영']!='-']
jeonnam_ed = jeonnam_ed[jeonnam_ed['목요일 운영']!='-']
jeonnam_ed = jeonnam_ed[jeonnam_ed['금요일 운영']!='-']
jeonnam_ed
```

```python
jeonnam_ed.to_csv('전남휴일지킴이약국.csv',encoding='cp949')
```

```python
jeonnam_ed = jeonnam_ed.reset_index()
```

```python
jeonnam_ed
```

```python
a = []
for i in range(len(jeonnam_ed)):
    p = jeonnam_ed['index'][i]
    a.append(p)

a
```

```python
jn_pharm_new = jn_pharm.drop(a)
jn_pharm_new
```

```python
jn_pharm_new.to_csv('전남약국_휴일지킴이제외.csv',encoding='cp949')
```
