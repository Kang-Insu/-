# 함평군 1km 고령

```python
hampyeong_older_1km = pd.read_csv('약국_함평_고령인구(1km).csv', encoding ='cp949')
hampyeong_older_1km
```

```python
h_older_1km_val_sum = hampyeong_older_1km.groupby(['약국명_y','CLUSTER_ID','좌표정보(X)','좌표정보(Y)','distance']).sum('val')
h_older_1km_val_sum = h_older_1km_val_sum[0:2]
h_older_1km_val_sum
```

```python
tmp = []
for i in h_older_1km_val_sum['val']:
    print(i)
    if i >= 0 and i <= 103:
        tmp.append(1)
    elif i >=104 and i <=206:
        tmp.append(2)
    elif i >= 207 and i <=309:
        tmp.append(3)
    elif i >=310 and i <= 412:
        tmp.append(4)
    else:
        tmp.append(5)
tmp
h_older_1km_val_sum['score'] = tmp
h_older_1km_val_sum
```

```python
h_older_1km_val_sum = h_older_1km_val_sum.reset_index()
h_older_1km_val_sum
```
